```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import kaleido

#df = px.data.gapminder()# Data Preparation
df_corr = df.corr() # Generate correlation matrix

fig = go.Figure()
fig.add_trace(
    go.Heatmap(
        x = df_corr.columns,
        y = df_corr.index,
        z = np.around(np.array(df_corr),decimals = 2),
        hovertemplate ="X: %{x}"+
                    "<br>Y: %{y}"+
                    "<br>Correlation Value: %{z}"+
                    "<extra></extra>",
        colorscale=[[0.0, 'rgb(141,160,203)'],
                    [0.5,'white'],
                    [1.0, 'rgb(252,141,98)']],
        hoverongaps = True,
        zmid=0,
        texttemplate="%{z}",
        colorbar=dict(
        title = "Correlation",
        titleside="top",
        orientation = 'h'

        )
    )
)

fig.update_layout(
    title='Correlation Matrix',
    xaxis_nticks=36,
    updatemenus=[
        dict(
            pad={"r": 10, "t": 10},
            x=0.1,
            xanchor="left",
            y=1.2,
            yanchor="top",

            
        ),
    ])
fig.layout.height = 900
fig.layout.width = 900

fig.show()
fig.write_html("HeatMap.html")
```

```python
x0 = df[df["Diabetes_binary"]==0]
x0 = x0["GenHlth"]
x1 = df[df["Diabetes_binary"]==1]
x1 = x1["GenHlth"]
```

```python
import plotly.graph_objects as go

import numpy as np
temp_color=px.colors.qualitative.Set2
fig = go.Figure()
fig.add_trace(go.Histogram(
    histfunc="count",
    x=x0,
    name='Without_Diabetes', # name used in legend and hover labels
    marker_color=temp_color[0],
    opacity=0.75
))
fig.add_trace(go.Histogram(
    histfunc="count",
    x=x1,
    name='Diabetes',
    marker_color=temp_color[1],
    opacity=0.75
))

fig.update_layout(
    plot_bgcolor = 'rgba(0,0,0,0)',
    title_text='GenHlth VS Diabetes', # title of plot
    xaxis_title_text='GenHlth Value', # xaxis label
    yaxis_title_text='Count', # yaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)
fig.update_yaxes(
                 showline=True, linewidth=1, gridcolor='lightgrey',
                 layer='below traces')
fig.show()
fig.write_html("GenHlth.html")
```
